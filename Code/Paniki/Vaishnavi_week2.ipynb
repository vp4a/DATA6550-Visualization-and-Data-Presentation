```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


df = pd.read_csv(r"C:\Users\vyshn\Downloads\WEEK2_DATA_6550\DATA\Titanic-Dataset.csv")
df.head()
```

### Good and Bad Scatter plot for Titanic Dataset

```python
survived_df = df[df["Survived"] == 1].copy()
died_df = df[df["Survived"] == 0].copy()

survived_df = survived_df.dropna(subset=["Fare"])
died_df = died_df.dropna(subset=["Fare"])

survived_df.reset_index(drop=True, inplace=True)
died_df.reset_index(drop=True, inplace=True)

plt.figure(figsize=(12, 10)) #plot size

plt.scatter(died_df.index, died_df["Fare"], color="red", alpha=0.5)#red dot for non-sue


sns.regplot(x=survived_df.index, y=survived_df["Fare"],
            scatter_kws={'alpha': 0.5, 'color': 'green'},
            line_kws={'color': 'blue'}) # Plot survived passengers (green dots) + trend line

plt.xlabel("Survival Count")
plt.ylabel("Fare Price")
plt.title("Scatter Plot: Survival vs Fare Price")
plt.grid(True, linewidth=2, linestyle=":")
plt.show()
```

In this scatter plot, I have clearly represented and differentiated between
passengers who survived and passengers who have not survived. I have
used green colour for the survivors and red for non-survivers to make a
distinct differentiation for the audience to have a clear understanding. Also,
the trend line helps to identify the patterns that show the fare price vs
survivor insights. The axes are labelled correctly to ensure the proper
understanding showing the relationship between survival and fare price.

```python
plt.figure(figsize=(12, 10))

plt.scatter(died_df.index, died_df["Fare"], color="yellow", alpha=0.1, s=5)
sns.regplot(x=survived_df.index, y=survived_df["Fare"],
            scatter_kws={'alpha': 1, 'color': 'purple', 's': 3},
            line_kws={'color': 'orange', 'linestyle': '--'})
plt.xlabel("Passenger Index", fontsize=5)
plt.ylabel("Fare Price ($)", fontsize=5)
plt.title("Scatter Plot", fontsize=8, color="red")

plt.grid(True, linewidth=2, linestyle=":")

plt.show()
```

In this Scatter plot, there are many unnecessary clutters making it difficult
for the data to interpret and it also lacks any distinction or proper data
scaling which can result in confusion on survival trends. Also, the
mislabeled axes and improper size of fonts make it hard to understand the
data. These all show how a poor design can lead to ethical concerns in
data presentation.

### Good and Bad Bar plot for Titanic Dataset

```python
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
df["family_size"] = df["SibSp"] + df["Parch"] + 1

sns.set_style("whitegrid")

plt.figure(figsize=(12, 10))
sns.barplot(
    data=df, x="family_size", y="Survived", ci=None, palette="Blues"
)
plt.xlabel("Family Size", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)
plt.title("Impact of Family Size on Survival", fontsize=14, fontweight="bold")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

This bar chart shows the impact of the size of the family on survival rates. I
have used a colour palette to represent different patterns, making viewers
to understand the data. Properly labelled axis and font sizes make it easy
to interpret the data, by following the consistency in accurately mapping the
data I have adhered to multiple visualisation principles ensuring no
misleading conclusions are drawn from the chart.

```python

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

family_survival = df.groupby('FamilySize')['Survived'].mean()
family_sizes = family_survival.index
survival_rates = family_survival.values

plt.figure(figsize=(12,10))
plt.bar(family_sizes, survival_rates, color=np.random.rand(len(family_sizes), 3), alpha=0.5)

plt.xticks(fontsize=2, rotation=45)
plt.yticks(fontsize=3)
plt.xlabel('Random Family Numbers', fontsize=3, color='blue')
plt.ylabel('Unclear % Survive?', fontsize=3, color='green')
plt.title('Impact of Family Size on Survival', fontsize=25, color='darkblue', style='oblique')

plt.legend([' '], loc='upper right', fontsize=20, frameon=True, fancybox=True, edgecolor="black")
plt.grid(False)
plt.show()
```

The above bar graph has many hindered and unreliable aspects, also due
to inaccurate labelling the accuracy of the graph is also highly affected.
Also, the colour palette makes it to difficult differentiate the values. No
proper legends, unreadable axes with small fonts and incorrect labels can
lead to a heavy misunderstanding of the survival rates. Moreover, the
design flaws can mislead the decision-making and create a false narrative
